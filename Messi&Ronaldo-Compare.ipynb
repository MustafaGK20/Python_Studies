```python
class footballer():
    def __init__(self, name, pace, shooting, passing, dribbling, defending, physical):
        
        self.name = name
        self.pace = pace
        self.shooting = shooting
        self.passing = passing
        self.dribbling = dribbling
        self.defending = defending
        self.physical = physical
    
    def talent_prepare(self):
        
        return ",".join([str(self.pace),
                         str(self.shooting),
                         str(self.passing),
                         str(self.dribbling),
                         str(self.defending),
                         str(self.physical),
                         str(self.pace)])
    
    def talent_visualize(self):
        
        graphic_url = "https://image-charts.com/chart"
        
        payload = {"chco" : "3092de",
                   "chd" : "t:" +self.talent_prepare(),
                   "chl" : "pace|shooting|passing|dribbling|defending|physical",
                   "chdl" : self.name,
                   "chdlp" : "b",
                   "chs" : "480x480",
                   "cht" : "r",
                   "chtt" : "Football Player Traits",
                   "chxl" : "0:|0|20|40|60|80|100",
                   "chxt" : "x",
                   "chxr" : "0,0.0,100.0",
                   "chm" : "B,AAAAAABB,0,0,0"}
        
        response = requests.post(graphic_url, data=payload)

        image = Image.open(BytesIO(response.content))
        image.show()
        
    def comparison(self, player_to_compare, ):
            
        graphic_url = "https://image-charts.com/chart"
        
        payload = {"chco" : "3092de|027182",
                   "chd" : "t:" +self.talent_prepare() + "|" + player_to_compare.talent_prepare(),
                   "chl" : "pace|shooting|passing|dribbling|defending|physical",
                   "chdl" : self.name + "|" + player_to_compare.name,
                   "chdlp" : "b",
                   "chs" : "480x480",
                   "cht" : "r",
                   "chtt" : "Football Player Traits",
                   "chxl" : "0:|0|20|40|60|80|100",
                   "chxt" : "x",
                   "chxr" : "0,0.0,100.0",
                   "chm" : "B,AAAAAABB,0,0,0|B,0073CFBB,1,0,0"}
        
        response = requests.post(graphic_url, data=payload)

        image = Image.open(BytesIO(response.content))
        image.show()
```

```python
messi = footballer("Messi", 85, 92, 91, 95, 38, 65)
ronaldo = footballer("Ronaldo", 89, 93, 81, 89, 35, 77)
```

```python
messi.talent_prepare()
```

```python
ronaldo.talent_prepare()
```

```python
ronaldo.comparison(messi)
```
